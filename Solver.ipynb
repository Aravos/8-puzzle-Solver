```python
import random

class nodes:
    def __init__(self,parent,mats,pos):  
        self.parent = parent   
        self.mats = mats
        self.pos = pos
        if(parent==None):
            self.depth = 0
        else:
            self.depth = parent.depth + 1
        
class puzzle:
    def __init__(self):
        self.curr = self.play()
        self.goal = {1: '1', 2: '0', 3: '3', 4: '4', 5: '2', 6: '6', 7: '7', 8: '5', 9: '8'}
    
    def printPath(self,root):
        if root != None:
            self.printPath(root.parent)
            self.printgrid(root.mats)
            print("")
            
    def blankpos(self,ini):
        for i in range(1,10):
            if(ini[i]=='0'):
                return i
    
    def printgrid(self,a):
        for j in range(1,10):
            print(a[j],end=" ")
            if(j%3==0):print(" ")
                
    def genmov(self,k,ini):
        child = {}
        if k not in [1,2,3]:
            a = ini.copy()
            temp = a[k]
            a[k] = a[k-3]
            a[k-3] = temp
            child[0]=a
        if k not in [9,8,7]:
            j = ini.copy()
            temp = j[k]
            j[k] = j[k+3]
            j[k+3] = temp
            child[1]=j
        if k not in [1,4,7]:
            b = ini.copy()
            temp = b[k]
            b[k] = b[k-1]
            b[k-1] = temp
            child[2]=b
        if k not in [3,6,9]:
            c = ini.copy()
            temp = c[k]
            c[k] = c[k+1]
            c[k+1] = temp
            child[3]=c
        return child
    
    def genmove(self,k,par):
        ini = par.mats
        child = []
        if k not in [1,2,3]:
            a = ini.copy()
            temp = a[k]
            a[k] = a[k-3]
            a[k-3] = temp
            m = self.blankpos(a)
            j = nodes(par,a,m)  
            child.append(j)
        if k not in [9,8,7]:
            j = ini.copy()
            temp = j[k]
            j[k] = j[k+3]
            j[k+3] = temp
            m = self.blankpos(j)
            j = nodes(par,j,m)  
            child.append(j)
        if k not in [1,4,7]:
            b = ini.copy()
            temp = b[k]
            b[k] = b[k-1]
            b[k-1] = temp
            m = self.blankpos(b)
            j = nodes(par,b,m)  
            child.append(j)
        if k not in [3,6,9]:
            c = ini.copy()
            temp = c[k]
            c[k] = c[k+1]
            c[k+1] = temp
            m = self.blankpos(c)
            j = nodes(par,c,m)  
            child.append(j)
        return child

    def BFS(self,OPEN,CLOSED,n):
        k = n.pos
        children = self.genmove(k,n)
        for i in children:
            flag = 0
            for j in OPEN:
                if(i.mats==j.mats):
                    flag=1
                    break
            if flag==0 and i.mats not in CLOSED:
                OPEN.append(i)
        return OPEN
    
    def DFS(self,OPEN,CLOSED,n):
        k = n.pos
        children = self.genmove(k,n)
        for i in children:
            if i.mats not in CLOSED:
                OPEN.insert(0,i)
        return OPEN
    
    def DFSL(self,OPEN,CLOSED,n,d):
        if(n.depth<d):
            return self.DFS(OPEN,CLOSED,n)
        else:
            if(OPEN!=[]):
                a = OPEN[0].depth
                m = d-a
                CLOSED = CLOSED[:m]
            return OPEN
    
    def iterD(self,d):
        q = self.__genSearch('3',d)
        if(q==0):
            return 0
        else:
            return 1
        
    def __genSearch(self,x,d=1):
        OPEN = []
        CLOSED = []
        G = []
        l = self.blankpos(self.curr)
        root = nodes(None,self.curr,l)
        OPEN.append(root)
        while(len(OPEN)>0):
            n = OPEN.pop(0)
            CLOSED.append(n.mats)
            if n.mats==self.goal:
                print("\nFOUND PATH\n")
                self.printPath(n)
                return 0
            else:
                if(x=='1'):
                    OPEN = self.BFS(OPEN,CLOSED,n)
                if(x=='2'):
                    OPEN = self.DFS(OPEN,CLOSED,n)
                if(x=='3'):
                    if(n.depth==d):
                        CLOSED.pop()
                    OPEN = self.DFSL(OPEN,CLOSED,n,d)
                if(x=='4'):
                    j = self.iterD(d)
                    if j==0:
                        return 0
                    else:
                        d=d+1
                        OPEN.append(root)
        return 1

    def start(self):
        x = input("Enter Algo to use:\n1.BFS\n2.DFS\n3.Depth Limited Search\n4.Iterative Depth Limited Search\n\nInput Choice: ")
        if(x=='3'):
            print("With 0 as root depth")
            d = int(input("Enter Depth: "))
            y = self.__genSearch(x,d)
        else:
            y = self.__genSearch(x)
        if y==1:
            print("\nPATH NOT FOUND\n")
        
        
    def play(self):
        curr = "123456780"
        ini ={}
        iterator = 1
        for i in range(3):
            l = []
            for i in range(3):
                ini[iterator] = curr[iterator-1]
                iterator+=1
        
        x = int(input("ENTER NUMBER OF MOVES TO MAKE: "))
        curr = ini
        prev = 10
        for i in range(0,x):
            k = self.blankpos(curr)
            child = self.genmov(k,curr)
            keysList = list(child.keys())
            nextmove = keysList[random.randint(0,len(keysList)-1)]
            while(1):
                if(nextmove==0):
                    if(prev==1):
                        nextmove = keysList[random.randint(0,len(keysList)-1)]
                    else:
                        break
                if(nextmove==1):
                    if(prev==0):
                        nextmove = keysList[random.randint(0,len(keysList)-1)]
                    else:
                        break
                if(nextmove==2):
                    if(prev==3):
                        nextmove = keysList[random.randint(0,len(keysList)-1)]
                    else:
                        break
                if(nextmove==3):
                    if(prev==2):
                        nextmove = keysList[random.randint(0,len(keysList)-1)]
                    else:
                        break
                
            prev = nextmove
            curr = child[nextmove]
        return curr
            
            
p1 = puzzle()
print("Current State")
p1.printgrid(p1.curr)
print("\nGoal State")
p1.printgrid(p1.goal)
print("")
n = 'y'
while (n=='y'):
    p1.start()
    n = input("Continue(y/n):")
print("EXIT")
```
